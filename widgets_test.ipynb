```
#Text Widgets
dbutils.widgets.text("SourcePath","")
```

```
#Text Widgets With text inside
dbutils.widgets.text("Path","/FileStore/tables")
```

```
#Dropdown Widget
dbutils.widgets.dropdown("Subject","payments",["payments","sales","orders","transactions"])
```

```
#combobox Widget
dbutils.widgets.combobox("travel","booking",["Scheduled","Track","Cancelled"])
```

```
#multiselect Widgets
dbutils.widgets.multiselect("columns","trainname",["trainname","traindate","trainsource","traindestination"])
```

```
#Getting the value from widget
loc = dbutils.widgets.get("columns")
print(loc)
```

```
#remove widget
dbutils.widgets.remove("SourcePath")
```

```
#to remove all
dbutils.widgets.removeAll()
```
