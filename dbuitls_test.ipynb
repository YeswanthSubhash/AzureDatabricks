```
# dbutils.fs.help()

# create a directory 'Samples' in dbfs
dbutils.fs.mkdirs('/FileStore/Samples')
```

```
# create a file and write to data into it.
dbutils.fs.put('/FileStore/Samples/test.txt',"Hi My Name is Yeswanth")
```

```
# display the file content
dbutils.fs.head('/FileStore/Samples/test.txt')
```

```
# List down the content in a directory
dbutils.fs.ls('/FileStore/Samples/')
```

```
# move file from one directory to another directory
dbutils.fs.mkdirs('/FileStore/Samples2/')
dbutils.fs.mv('/FileStore/Samples/test.txt','/FileStore/Samples2/')
dbutils.fs.ls('/FileStore/Samples2/')
```

```
#copy file from one location to another
dbutils.fs.cp('/FileStore/Samples2/test.txt','/FileStore/Samples/')
dbutils.fs.ls('/FileStore/Samples/')
```

```
# remove files and directories
dbutils.fs.rm('/FileStore/Samples2/test.txt')
```

```
# to mount the paths
dbutils.fs.mounts()
```

```
# to unmounts
dbutils.fs.unmounts()
```

```
#install a package
dbutils.library.installPyPI("tensorflow")
```
